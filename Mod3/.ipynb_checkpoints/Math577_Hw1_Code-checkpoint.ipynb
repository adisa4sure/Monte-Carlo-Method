### Volume of d-dimensional Unit Ball

```python
import numpy as np

def generate_random_point(d):
    return np.random.uniform(-1, 1, d)

def is_inside_unit_ball(point):
    return np.linalg.norm(point) <= 1

def estimate_volume(d, num_points):
    inside_count = 0

    for _ in range(num_points):
        point = generate_random_point(d)
        if is_inside_unit_ball(point):
            inside_count += 1

    volume_estimate = (inside_count / num_points) * (2 ** d)
    return volume_estimate

# Example usage
point = [10**i for i in range(2, 10)]
volume_estimates = []
for num_points in point:
    d = 3
    volume_estimate = estimate_volume(d, num_points)
    volume_estimates.append(volume_estimate) 
    
print(f"Volumes of {d}-dimensional Unit Ball for diff number N points: {volume_estimates}")
```

```python
# Plotting the graph of true volume and estimated volumes for comparison.
import matplotlib.pyplot as plt
import math

true_volume = ((np.pi)**(d/2))/(math.gamma((d/2)+1)) # Volume of d_dim unit sphere

plt.plot(np.log10(point), true_volume*np.ones_like(point), label='True Volume', linestyle='--', color='blue')
plt.plot(np.log10(point), volume_estimates, marker='o', label='volume_estimates', linestyle='-', color='red')
plt.title('The Plot of True Volume and Estimated volumes')
plt.xlabel('Num of Points in Power of 10')
plt.ylabel('Volumes')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Computing and plotting error difference between the true volume and estimated
# for different number of points
error_difference = np.abs(true_volume - np.array(volume_estimates))

plt.plot(np.log10(point), error_difference, marker='o')
plt.title('Volumes Error Difference vs Number of Points')
plt.xlabel('Num of Points in Power of 10')
plt.ylabel('Error Difference')
plt.grid(True)
plt.show()
```

```python
# Computing volume of different dimension of a unit sphere
dimensions = [3,4,5,6,7]
volume_estimates2 = []
for d in dimensions:
    num_points = 10**8
    volume_estimate = estimate_volume(d, num_points)
    volume_estimates2.append(volume_estimate) 
    
print(f"Volumes of different d-dimensional Unit Ball for {num_points} points: {volume_estimates2}")
```

```python
# Computing and plotting error difference between the true and estimated
# Volumes for different dimensions
true_volume2 = np.array([((np.pi)**(d/2))/(math.gamma((d/2)+1)) for d in dimensions]) # Volume of d_dim unit sphere
error_difference2 = np.abs(true_volume2 - np.array(volume_estimates2))

plt.plot(dimensions, error_difference2, marker='o')
plt.title('Volumes Error Difference vs Dimensions')
plt.xlabel('Dimensions')
plt.ylabel('Error Difference')
plt.grid(True)
plt.show()
```

### Surface Area of d-dimensional Unit Ball

```python
import random
import math

def monte_carlo_unit_ball_surface_area(dimension, num_points):
    points_inside_unit_ball = 0

    for _ in range(num_points):
        point = [random.uniform(-1, 1) for _ in range(dimension)]
        distance_squared = sum(x**2 for x in point)

        if distance_squared <= 1:
            points_inside_unit_ball += 1

    ratio = points_inside_unit_ball / num_points
    unit_ball_surface_area = ratio * 2*d*(2**(d-1))

    return unit_ball_surface_area, ratio

# Example usage:
d = 3
num_points = 1000000
result = monte_carlo_unit_ball_surface_area(d, num_points)
print(f"Estimated Ratio of {d}-dimensional Unit Ball: {result[1]}")
print(f"Estimated Surface Area of {d}-dimensional Unit Ball: {result[0]}")
```

```python
# Computing Surface Area of a unit ball for different number of points
point2 = [10**i for i in range(2, 10)]
surface_estimates = []
for num_points in point2:
    d = 3
    result = monte_carlo_unit_ball_surface_area(d, num_points)
    surface_estimates.append(result[0])
print(f"Estimated Surface Area of {d}-dimensional Unit Ball for number N points: {surface_estimates}")
```

```python
# Plotting the graph of true Surface Area and estimated Surface Areas for comparison.
true_surface_area = (2*(np.pi)**(d/2))/(math.gamma(d/2)) # Surface Area of d_dim unit sphere

plt.plot(np.log10(point2), true_surface_area*np.ones_like(point2), label='True S.Area', linestyle='--', color='blue')
plt.plot(np.log10(point2), surface_estimates, marker='o', label='surface_estimates', linestyle='-', color='red')
plt.title('The Plot of True and Estimated Surface Areas')
plt.xlabel('Num of Points in Power of 10')
plt.ylabel('Surface Areas')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Computing and plotting error difference between the true and estimated
# Surface Areas for different number of points
error_differencee3 = np.abs(true_surface_area - np.array(surface_estimates))

plt.plot(np.log10(point2), error_differencee3, marker='o')
plt.title('Surface Area Error Difference vs Number of Points')
plt.xlabel('Num of Points in Power of 10')
plt.ylabel('Error Difference')
plt.grid(True)
plt.show()
```

```python
# Computing Surface Area of different dimension of a unit sphere
dimensions3 = [3,4,5,6,7]
surface_estimates3 = []
for d in dimensions3:
    num_points = 10**8
    surface_estimate = monte_carlo_unit_ball_surface_area(d, num_points)
    surface_estimates3.append(surface_estimate) 
    
print(f"Surface Area of different d-dimensional Unit Ball for {num_points} points: {surface_estimates3}")
```

```python
# Computing and plotting error difference between the true and estimated
# Surface Area for different dimensions
true_surface_area3 = np.array([(2*(np.pi)**(d/2))/(math.gamma(d/2)) for d in dimensions3]) # Surface Area of d_dim unit sphere
surface_estimates3 = [t[0] for t in surface_estimates3]
error_difference3 = np.abs(true_surface_area3 - surface_estimates3)
plt.plot(dimensions3, error_difference3, marker='o')
plt.title('Surface Area Error Difference vs Dimensions')
plt.xlabel('Dimensions')
plt.ylabel('Error Difference')
plt.grid(True)
plt.show()
```
